신경망 학습이란 훈련 데이터로부터 가중치 매개변수의 최적값을 자동으로 획득하는 것을 뜻한다. 신경망이 학습할 수 있도록 해주는 지표인 손실 함수를 소개한다. 손실 함수의 이름 그래도 손실이 가장 적을 경우의 가중치가 우리가 학습시킨 결과가 될 것이다. 이번 장에서는 손실 함수의 값을 가급적 작게 만드는 기법으로 함수의 기울기를 활용하는 경사법을 소개합니다.

신경망의 특징은 데이터를 보고 학습을 할 수 있다는 점이다. 데이터에서 학습한다는 것은 가중치 매개변수의 값을 데이터를 보고 자동으로 결정한다는 뜻이다. 따라서 이번 장에서는 신경망 학습에 대해 설명하고 3장에서 한 MNIST의 가중치를 직접 구해보자.

# 4.1.1 데이터 주도 학습

기계학습은 데이터가 생명이다. 데이터에서 답을 찾고 데이터에서 패턴을 발견하고 데이터로 이야기를 만드는, 그것이 바로 기계학습이다. 이처럼 기계학습은 데이터가 이끄는 접근 방식 덕에 사람 중심 접근에서 벗어날 수 있었다. 추가적으로 신경망과 딥러닝은 기존 기계학습에서 사용하던 방법보다 사람의 개입을 더욱 배제할 수 있게 해주는 중요한 특성을 지녔다.

<img src="images/4_01.png" height="100px" width="300px">
위의 그림을 보면 숫자 5를 판별하기 위한 알고리즘을 어디서 부터 짜야할지, 어떻게 짜야할지 고민이 되는데, 사실상 밑바닥부터 설계하기 보다는 주어진 데이터를 활용하고 싶을 것이다. 그런 방법의 하나로 이미지에서 "특징(feature)"을 추출하고 그 특징의 패턴을 기계학습 기술로 학습하는 방법이 있다. 여기서 말하는 특징은 입력 데이터에서 본직적인 데이터를 정확하게 추출할 수 있도록 설계된 변환기를 가리킨다. 이미지의 특징은 보통 벡터로 기술한다.

이와 같은 기계학습에서는 모아진 데이터로부터 규칙을 찾아내는 역할을 "기계"가 담당한다. 이는 "무"로 부터 알고리즘을 설계하는 것보다 효율이 높아 문제를 해결해야 하는 사람의 부담도 덜어준다. 다만, 이미지를 벡터로 변환할 때 사용하는 특징은 여전히 "사람"이 설계하는 것이기에, 적합한 특징을 쓰지 않으면 좀처럼 좋은 결과를 얻을 수 없다.

여기서 볼 수 있듯이 신경망은 앞서 소개한 두 방식과 다른게 사람의 개입 없이 이미지를 "있는 그대로" 학습한다. 이처럼 신경망의 이점은 모든 문제를 같은 맥락에서 풀 수 있다는 점이다. "5"를 인식하는 문제든, "개"를 인식하는 문제든, 세부사항과 관계없이 신경망은 주어진 데이터를 있는 그대로 학습하고, 주어진 문제의 패턴을 발견하려 시도한다.

# 4.1.2 훈련 데이터와 시험 데이터

기계학습 문제는 데이터를 훈련 데이터와 시험 데이터로 나누어 학습과 실험을 수행하는 것이 일반적이다. 우선은 훈련 데이터만을 사용해 최적의 가중치들을 찾아내고, 이를 시험 데이터를 이용해 훈련된 모델을 평가하는 것이다. 이는 모델을 만드는 목적과 관련되어 있는데, 우리는 어디서든 활용가능한 범용적인 모델을 설계하는 것이 목적이기 때문에 훈련 데이터와 시험 데이터를 분리하게 되는 것이다. 이렇게 데이터셋을 쪼개지 않으면, 훈련 데이터에만 과도하게 최적화 되어 이후 새로운 데이터에 대해서는 잘못된 결과를 출력하게 되는 "오버피팅(overfitting)"을 발생시킬 수 있기 때문이다.
